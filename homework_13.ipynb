```python
# задание 1
import pandas as pd
```

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')
ratings.head()
```

```python
new_ratings = ratings['userId'].value_counts().reset_index()
new_ratings.head()
```

```python
users_count = pd.DataFrame(new_ratings)
users_count.columns = ['user', 'value']
users_count.head()
```

```python
over_100 = users_count[users_count['value'] > 100]
over_100.head()
```

```python
rating100 = ratings[ratings['userId'].isin(over_100.user.tolist())]
rating100.head()
```

```python
results = rating100.groupby('userId').agg({'timestamp': ['min', 'max']})
results.head()
```

```python
results['lifetime'] = results['timestamp']['max'] - results['timestamp']['min']
results.head()
```
